# Task 5: Use the appropriate tests to answer the questions provided.

## Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

Hypothesis Statement:
<br>
H_0: the mean value of MEDV are the same for boudned by the Charles river or not 
<br>
H_1: the mean value of MEDV are different for boudned by the Charles river or not 

```python
scipy.stats.ttest_ind(boston_df[boston_df["CHAS"] == 0]["MEDV"],boston_df[boston_df["CHAS"] == 1]["MEDV"]) 
```

Since p value < 0.05, we reject the hypothesis.  <br>
So the MEDV is different for boudned by the Charles river or not.

## Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

Hypothesis Statement:
<br>
H_0: the mean value of MEDV are the same for different groups of population.
<br>
H_1: the mean value of MEDV are not all the same for different groups of population.

```python
scipy.stats.f_oneway(boston_df[boston_df["age_group"] == '35 years and younger']["MEDV"],
                     boston_df[boston_df["age_group"] == 'between 35 and 70 years']["MEDV"],
                    boston_df[boston_df["age_group"] =='70 years and older']["MEDV"])
```

Since p value < 0.05, we reject the hypothesis.
So the MEDV is different for different age groups.

## Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

Hypothesis Statement:
<br>
H_0: NOX is not correlated with INDUS
<br>
H_1: NOX is correlated with INDUS

```python
Pearson_Coefficient_value, P_value=scipy.stats.pearsonr(boston_df["NOX"],boston_df["INDUS"])
Pearson_Coefficient_value, P_value
```

Since p value < 0.05, we reject the hypothesis.
So the NOX is correlated with INDUS.

## What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

Hypothesis Statement:
<br>
H_0: DIS is not correlated with MEDV
<br>
H_1: DIS is correlated with MEDV

```python
y=boston_df["MEDV"]
x=boston_df["DIS"]
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
model.summary()
```

Pearson_R_pvalue = 1.21e-08, R-squred = 0.062

```python
Pearson_R_correlation_coefficient=np.sqrt(0.062)
Pearson_R_correlation_coefficient
```

Since Pearson_R_pvalue < 0.05, we reject the hypothesis. So the DIS is correlated with MEDV.
<br>
Since Pearson_R_correlation_coefficient is 0.249, so DIS has relatively weak impact on MEDV.
